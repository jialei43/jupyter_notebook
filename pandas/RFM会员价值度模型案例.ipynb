## 数据预处理

```python
# 遍历,获取每张表的数据
for i in sheet_name[0:-1]:
    # 删除缺失值
    execl_data[i] = execl_data[i].dropna()
    # 过滤大于1的订单
    execl_data[i] = execl_data[i][execl_data[i]['订单金额']>1]
    # 提取提交日期的最大值
    execl_data[i]['max_year_date'] = execl_data[i]['提交日期'].max()
    
```

```python
for e in sheet_name:
    print(execl_data[e].info())
```

```python
# 合并sheet页数据
df_merge = pd.concat(list(execl_data.values())[0:-1],ignore_index=True)
df_merge.tail()
```

```python
# 验证订单金额小于1的订单是否被过滤
count = len(df_merge[df_merge['订单金额'] <= np.nan])
count
```

```python
# 为了好区分,df兑现新增year列
df_merge['year'] = df_merge['提交日期'].dt.year
df_merge
```

```python
# 计算最大日期和所有订单日期的差值,作为RFM模型中的R值(时间间隔)
df_merge['date_interval'] = df_merge['max_year_date'] - df_merge['提交日期']
# 将时间间隔转换为天数
df_merge['date_interval'] = df_merge['date_interval'].dt.days
```

```python
df_merge
```

```python
# 基于year 和会员id进行分组,计算 RFM模型中的R,F和M值
RFM_model = df_merge.groupby(['会员ID','year'],as_index=False).agg(
    r = ('date_interval','min'),
    f = ('订单号','count'),
    m = ('订单金额','sum')
)
RFM_model
```

```python
RFM_model.iloc[:,2:].describe().T
```

```python
# 思路一:系统自动划分
# pd.cut(RFM_model['r'],3,include_lowest=True).unique()
```

```python
# 思路二:手动划分
r_bins = [-1,79,255,365]
f_bins = [0,2,5,130]
m_bins = [1,69,1199,206252]

# cut 的bins是左开又闭的
RFM_model['r_lable'] = pd.cut(RFM_model['r'],bins=r_bins,labels=[i for i in range(len(r_bins)-1,0,-1)])
RFM_model['f_lable'] = pd.cut(RFM_model['f'],bins=f_bins,labels=[i for i in range(1,len(f_bins),1)])
RFM_model['m_lable'] = pd.cut(RFM_model['m'],bins=m_bins,labels=[i for i in range(1,len(f_bins),1)])
RFM_model
```

```python
RFM_model['r_lable'] = RFM_model['r_lable'].astype(str)
RFM_model['f_lable'] = RFM_model['f_lable'].astype(str)
RFM_model['m_lable'] = RFM_model['m_lable'].astype(str)

RFM_model['rfm_group']=RFM_model['r_lable']+RFM_model['f_lable']+RFM_model['m_lable']
RFM_model
```

```python
# 5. 数据可视化.
# 5.1 准备可视化的数据, 即: rfm_group(分组结果评分), year(统计年份), number(评分个数)
display_data = RFM_model.groupby(['rfm_group', 'year'], as_index=False).agg({'会员ID': 'count'})
display_data
# # 5.2 修改列名.
display_data.columns = ['rfm_group', 'year', 'number']
# # 细节: 把number列的类型 -> int类型.
display_data['number'] = display_data['number'].astype(int)
# display_data
# 
# # 5.3 绘制图形
# 颜色池
range_color = ['#313695', '#4575b4', '#74add1', '#abd9e9', '#e0f3f8', '#ffffbf',
               '#fee090', '#fdae61', '#f46d43', '#d73027', '#a50026']
# 
#找出会员数最大的值,用于后面设置颜色映射范围
range_max = int(display_data['number'].max())
c = (
    Bar3D()  #初始化D3柱状图
    .add(
        "",  #图例 不显示图例
        [d.tolist() for d in display_data.values],  #数据
        xaxis3d_opts=opts.Axis3DOpts(type_="category", name='分组名称'),  #x轴数据类型，名称，rfm_group
        yaxis3d_opts=opts.Axis3DOpts(type_="category", name='年份'),  #y轴数据类型，名称，year
        zaxis3d_opts=opts.Axis3DOpts(type_="value", name='会员数量'),  #z轴数据类型，名称，number
    )
    .set_global_opts(  # 全局设置
        visualmap_opts=opts.VisualMapOpts(max_=range_max, range_color=range_color),  #设置颜色，及不同取值对应的颜色
        title_opts=opts.TitleOpts(title="RFM分组结果"),  #设置标题
    )
)
c.render()  # 数据保存到本地的网页中.
# c.render_notebook() #在notebook中显示
```
