```python
import pandas as pd
```

```python
df = pd.read_csv("./data/movie.csv")
```

```python
#（1）按照导演进行分组，计算每位导演票房收入（gross列）总和
df_group = df.groupby("director_name", as_index=False).agg(
    {"gross": "sum"}
)
df_group
```

```python
# （2）对上一步结果进行排序，求出票房前十的导演名称
df_group.gross.sort_values(ascending=False).head(10)
```
