## 作业2：学生考试成绩

```python
import numpy as np

# 创建成绩数据（单位：分）
scores_data = {
    '学生ID': ['S001', 'S002', 'S003', 'S004', 'S005'],
    '数学': [85, 92, 76, 88, 95],
    '英语': [78, 85, 82, 90, 88],
    '物理': [90, 87, 91, 79, 83]
}

# 转换为ndarray（以学生ID为索引）
scores_ndarray = np.array([
    [85, 78, 90],
    [92, 85, 87],
    [76, 82, 91],
    [88, 90, 79],
    [95, 88, 83]
], dtype=np.float64)

# 设置学生ID为行索引（实际操作中可通过花式索引实现）
student_ids = np.array(['S001', 'S002', 'S003', 'S004', 'S005'])
subjects = np.array(['数学', '英语', '物理'])
```

```python
# 查看成绩数组的维度、形状、元素类型及总个数
print(scores_ndarray.ndim)
print(scores_ndarray.shape)
print(scores_ndarray.dtype)
print(scores_ndarray.size)
```

```python
# 将成绩数组转换为整数类型，并查看每个元素的字节大小
scores_ndarray.astype(dtype=np.int64)
print(scores_ndarray)
print(scores_ndarray.itemsize)
```

```python
# 计算所有学生的数学平均分、物理总分及英语最高分
print(scores_ndarray[0:, 0].mean())
print(scores_ndarray[0:, 2].sum())
print(scores_ndarray[0:, 1].max())
```

```python
# 对物理成绩进行排序，并找到最低分的索引
print(np.argmin(scores_ndarray[:, 2]))
```

```python
# 使用where函数将 90 分以上的成绩标记为 'A'，其余为 'B'
np.where(scores_ndarray > 90, 'A', 'B')
```

```python
# 计算每个学生的总分（行方向求和）与每科的平均分（列方向求平均）
print(scores_ndarray.sum(axis=1))
print(scores_ndarray.mean(axis=0))
```

```python
# 使用cumsum计算每个学生的累计得分
np.cumsum(scores_ndarray, axis=1)
```
