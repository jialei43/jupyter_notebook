## 第三步:完成具体绘制

```python
plt.figure(figsize=(73,32),dpi=80)
x = [i for i in range(1,366)]
y = np.random.randint(-40,40,365)
plt.plot(x,y)
plt.xticks(x[::5],x[::5],rotation=45)
# plt.yticks(y[::5],y[::5],rotation=45)
plt.xlabel('时间',fontsize=100)
plt.ylabel('气温',fontsize=100)
plt.show()
```

```python
# 创建365天的时间序列
days = np.arange(1, 366)

# 假设气温随季节变化，用正弦波模拟气温波动
# 假设年温度变化从0到40度，正弦波周期为365天（一年），幅度为20度
temperature = 20 * np.sin(2 * np.pi * days / 365) + 20

# 创建图形
plt.figure(figsize=(10, 6))

# 绘制气温变化曲线
plt.plot(days, temperature, label="Temperature", color='b', linewidth=2)

# 添加标题和标签
plt.title("Temperature Change Over 365 Days", fontsize=16)
plt.xlabel("Day of the Year", fontsize=12)
plt.ylabel("Temperature (°C)", fontsize=12)

# 添加网格
plt.grid(True)

# 显示图例
plt.legend()

# 显示图形
plt.tight_layout()
plt.show()
```

```python
# 一个画布中绘制俩个折现 俩次plot
plt.figure(figsize=(10, 6))

x = range(60)
y_beijing = np.random.randint(1,3,60)
y_shanghai = np.random.randint(15,18,60)

plt.plot(x, y_beijing, label="北京", color='red', linestyle='--', linewidth=1)
plt.plot(x, y_shanghai, label="上海", color='blue', linestyle='-', linewidth=1)

x_ticks_table =[f'11点{e}分' for e in range(60)]
y_ticks_table = range(-10,31)

plt.xticks(x[::5],x_ticks_table[::5])
plt.yticks(y_ticks_table[::5])
# 添加标题和标签
plt.title("北京上海1小时内气温变化", fontsize=16)
plt.xlabel("时间", fontsize=12)
plt.ylabel("气温 (°C)", fontsize=12)
# 添加网格
plt.grid(linestyle='--', alpha=0.5)
# 添加图例
plt.legend()

# 保存图片
plt.savefig('./beijing_shanghai.png', dpi=300)

plt.tight_layout()
plt.show()
```

```python
# 同一张画布,不同坐标系
fig,axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 6), dpi=300)

x = range(60)
y_beijing = np.random.randint(1,3,60)
y_shanghai = np.random.randint(15,18,60)

# 注意:先要先创建坐标系,再绘制数据
axes[0].plot(x, y_beijing, label="北京", color='red', linestyle='--', linewidth=1)
axes[1].plot(x, y_shanghai, label="上海", color='blue', linestyle='-', linewidth=1)

x_ticks_table =[f'11点{i}分' for i in range(60)]
y_ticks_table = range(-10,31)

# 添加x,y轴刻度及标签
axes[0].set_xticks(x[::5],x_ticks_table[::5])
axes[0].set_yticks(y_ticks_table[::5])

axes[1].set_xticks(x[::5],x_ticks_table[::5])
axes[1].set_yticks(y_ticks_table[::5])

# 添加标题
axes[0].set_title("北京1小时内气温变化", fontsize=16)
axes[1].set_title("上海1小时内气温变化", fontsize=16)

# 添加行标签
axes[0].set_xlabel("时间", fontsize=12)
axes[1].set_xlabel("时间", fontsize=12)

# 添加列标签
axes[0].set_ylabel("气温 (°C)", fontsize=12)
axes[1].set_ylabel("气温 (°C)", fontsize=12)

# 添加图例
axes[0].grid(linestyle='--', alpha=0.5)
axes[1].grid(linestyle='--', alpha=0.5)

# 添加图例
axes[0].legend()
axes[1].legend()

"""
plt.tight_layout() 是 matplotlib 中的一个方法，用于自动调整子图（subplot）之间的间距，使得图形的各个部分（标题、坐标轴标签、刻度标签等）不会重叠，并且能够更好地适应图像区域。
"""
plt.tight_layout()
plt.show()

```

```python
plt.style.use('_mpl-gallery')  # 设置matplotlib的绘图样式，采用'mpl-gallery'样式，使图表更美观

# 生成数据
np.random.seed(3)  # 设置随机数种子，使得每次生成的随机数相同，方便调试和复现结果
x = 4 + np.random.normal(0, 2, 24)  # 生成24个均值为4，标准差为2的正态分布随机数，赋值给x
y = 4 + np.random.normal(0, 2, len(x))  # 生成24个均值为4，标准差为2的正态分布随机数，赋值给y

# 设置散点图的点的大小和颜色
sizes = np.random.uniform(15, 80, len(x))  # 生成24个大小在15到80之间的均匀分布随机数，作为散点的大小
colors = np.random.uniform(15, 80, len(x))  # 生成24个颜色值，范围在15到80之间，用作散点的颜色

# 创建图表
fig, ax = plt.subplots()  # 创建一个图表fig和一个坐标轴ax，用于绘图

# 绘制散点图
ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)  # 绘制散点图，x为横坐标，y为纵坐标，s为点的大小，c为点的颜色，vmin和vmax设置颜色值的范围

# 设置坐标轴范围和刻度
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),  # 设置x轴的范围为0到8，刻度为1到7
       ylim=(0, 8), yticks=np.arange(1, 8))  # 设置y轴的范围为0到8，刻度为1到7

# 显示图表
plt.show()  # 显示图表
```

```python
import matplotlib.pyplot as plt  # 导入matplotlib库中的pyplot模块，用于绘图
import numpy as np  # 导入numpy库，用于数值计算

plt.style.use('_mpl-gallery')  # 设置matplotlib的绘图样式，采用'mpl-gallery'样式，使图表更美观

# 生成数据
x = np.linspace(0, 10, 100)  # 创建一个从0到10，包含100个点的等间距的数组x，用于绘制x轴数据
y = 4 + 1 * np.sin(2 * x)  # 使用sin函数生成y轴数据，y = 4 + sin(2x)（一个正弦波函数）
x2 = np.linspace(0, 10, 25)  # 创建一个从0到10，包含25个点的等间距的数组x2，作为第二组数据
y2 = 4 + 1 * np.sin(2 * x2)  # 使用sin函数生成y2轴数据，y2 = 4 + sin(2x2)（另一组正弦波函数）

# 创建图表
fig, ax = plt.subplots()  # 创建一个图表fig和一个坐标轴ax，用于绘图

# 绘制不同样式的线图
ax.plot(x2, y2 + 2.5, 'x', markeredgewidth=2)  # 绘制x2和y2数据点，'x'表示点的形状是交叉，markeredgewidth=2设置点的边框宽度
ax.plot(x, y, linewidth=2.0)  # 绘制x和y数据点，线条宽度为2.0
ax.plot(x2, y2 - 2.5, 'o-', linewidth=2)  # 绘制x2和y2数据点，'o-'表示点和线的组合，线条宽度为2

# 设置坐标轴范围和刻度
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),  # 设置x轴的范围为0到8，刻度为1到7
       ylim=(0, 8), yticks=np.arange(1, 8))  # 设置y轴的范围为0到8，刻度为1到7

# 显示图表
plt.show()  # 显示图表
```
