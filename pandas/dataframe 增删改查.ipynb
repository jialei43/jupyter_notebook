# dataframe 增删改查

## 增

```python
import pandas as pd

df = pd.read_csv('./data/1960-2019全球GDP数据.csv', encoding='GBK')
df2 = df.head().copy()
df2
```

### 方式一: df['列名']=值 添加列

```python
# 方式一:df2['c1'] = 23 添加列
df2['c1']= 23
df2
```

```python
df2['c2'] = ['huahua', 'xiaoming', 'xiaohong', 'xiaoli', 'xiaobai']
df2
```

```python
df2['c3'] = df2.year*2
df2
```

```python
def f():
    return 2500
df2['c4'] = f()
df2
```

```python
df2['c5']= (lambda :2500)()
df2
```

### 方式二: df.assign(列名=值)

```python
df3 = df[:5].df4()
# df3.assign(c1=23)
df3
```

#### 添加单列 df.assign(列名=值)

```python
# 增加单列
df3.assign(c1=23)
```

#### 添加多列 df.assign(列名=值,列名=值)

```python
df3.assign(
    c2='huahua', 
    c3=(lambda :2500)(),
    c4 = f(),
    c5 =['huahua', 'xiaoming', 'xiaohong', 'xiaoli', 'xiaobai'],
    c6 = df3.year*2
)
```
