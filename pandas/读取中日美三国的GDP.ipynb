# 文件读取、数据可视化

## 1.1 导包

```python
#导包
import pandas as pd
# import matplotlib
# matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
```

## 1.2 读取文件

```python
df = pd.read_csv("./data/1960-2019全球GDP数据.csv", encoding="GBK")
print(df)
print('-'*34)
print(df[df.country=='中国'].head())
```

## 1.3 绘制折线图

```python
df2 = df.copy()
df2.set_index("year", inplace=True)
df2[df2.country=="中国"].GDP.plot()
```

## 1.4 绘制多国折线图

```python
df2[df2.country=="中国"].GDP.plot()
df2[df2.country=="美国"].GDP.plot()
df2[df2.country=="日本"].GDP.plot()
```

## 1.5 绘制中文折线图

```python
# 设置中文字体（macOS 兼容版本）
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS', 'PingFang SC', 'Heiti SC']
plt.rcParams['axes.unicode_minus'] = False

# 对指定列名进行修改
df2[df2.country=="中国"].rename(columns={"GDP":"中国GDP"}).中国GDP.plot(legend=True)
df2[df2.country=="美国"].rename(columns={"GDP":"美国GDP"}).美国GDP.plot(legend=True)
df2[df2.country=="日本"].rename(columns={"GDP":"日本GDP"}).日本GDP.plot(legend=True)
```
