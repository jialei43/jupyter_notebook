## 索引操作

###  场景1:根据行索引进行操作数据

```python
# 场景1:根据行列索引进行操作数据,必须是先列后行
df["open"]["2018-02-22"]
```

###   场景2:结合loc 根据 行索引和列索引来获取数据

```python
#场景2 :结合loc 根据 行索引和列索引来获取数据
df.loc["2018-02-22", "open"]
```

### 场景3 :结合loc 根据行索引进行切片和列索引的集合 获取数据

```python
### 场景3 :结合loc 根据行索引和列索引通过切片来获取数据
df.loc["2018-02-27":"2018-02-14", ["open", "high"]]
```

###  场景4 :结合iloc 根据行列索引进行切片(不包尾) 获取数据

```python
df.iloc[0:3, 0: 2]
```

###  场景5 :结合at 根据行索引值和列索引值获取数据

```python
#  场景5 :结合at 根据行索引值和列索引值获取数据
# .at[] 是基于标签（行索引和列标签）进行访问和赋值，速度较快，适合单个元素的修改。
df.at["2018-02-22", "open"]
```

###   场景6 :结合iat 根据行索引下标和列索引下标获取数据

```python
df.iat[8, 5]
```

### 总结: loc,iloc,at,iat  区别

> 总结: 
> 1. loc和iloc都是基于标签进行访问和赋值，速度较at慢，适合多维数据的访问。
> 2. at是基于标签进行访问和赋值，速度较loc和iloc快，适合单个元素的访问。
> 3. iat是基于索引进行访问和赋值，速度较at和loc快，适合单个元素的访问。

#### apply() 自定义函数运算

```python
#自定义函数 my_ func 接收一列数据,计算该列的最大值和最小值,并计算差值并返回
def  my_func(x):
    return x.max() - x.min()

df[["open","close"]].apply(my_func, axis=1)  # 行运算
```

```python
df[["open","close"]].apply(my_func, axis=0) #  列运算
```

```python
df[["open","close"]].apply(lambda x: x.max() - x.min()) #默认,列运算
```
