## fillna()函数 将Nan替换为指定值

```python
df = pd.read_csv("data/movie.csv")
df.info()
```

```python
df1 = df.copy()
df1.fillna(df1['Runtime (Minutes)'].mean(), inplace=True)
df1.info()
```

```python
df2 = df.copy()
df2['Revenue (Millions)'].fillna(df2['Revenue (Millions)'].mean(), inplace=True)
df2.info()
```
