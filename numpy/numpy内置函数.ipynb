# 4 np.sort() 排序函数

## 4.1 np.sort()会返回一个新的ndarray对象,不修改原ndarray对象

```python
reshape = np.random.randint(0,25, size=(5,5))
print(np.sort(reshape))
```

## 4.2 ndarray.sort() 对象调用sort()会修改原ndarray对象

```python
print(f'reshape 排序前:{reshape}')
reshape.sort()
print(f'reshape 排序后:{reshape}')
```

# 5 矩阵运算

## 5.1 numpy基础运算

```python
# numpy 基本运算 
# 前提: 两个矩阵的行数和列数必须完全相同(矩阵加减)
arr1=np.array([10,20,30,40])
arr2=np.arange(4)
print(arr1)
print(arr2)
#减法运算 
print(arr1-arr2)  # 对应位置元素做减法运算(行列数相同)
#加法运算
print(arr1+arr2)  # 对应位置元素做加法运算(行列数相同)
#乘法运算
print(arr1*arr2)  # 对应位置元素做乘法运算(行列数相同)
#除法运算
print(arr1/arr2)
```

# 5.2 矩阵乘法

```python
# 矩阵乘法  
# 定义数组
arr1=np.array([[1,2,3],
         [4,5,6]])

arr2=np.array([[1,2,3],
         [4,5,6]])

# print(f"arr1:{arr1} , {arr1.shape}")
# print(f"arr2:{arr2}, {arr1.shape}")
#矩阵乘法   行列数一致   哈达玛积
# 哈达玛积一种点对点的乘法运算 , 仅限于形状完全相同的矩阵(或者向量)
# print(arr1*arr2)
# print(np.multiply(arr1,arr2))
# 矩阵乘法  ->行列数不一致
arr3=np.array([[1,2,3],
         [4,5,6]])   # 2,3
arr4=np.array([
    [9,8],
    [6,5],
    [3,2]
    ]
)       #3,2
# print(arr3*arr4)                #报错,行列数不一致(乘法适合行列数一致场景) 
# print(np.multiply(arr3,arr4))   #报错,行列数不一致(乘法适合行列数一致场景) 
# 
#dot
# print(np.dot(arr3,arr4))   
# print(arr3.dot(arr4))       #  效果同上   
print(arr3@arr4)

#矩阵相乘 行列数不一致: A矩阵: 2行3列   B:矩阵 3行2列    A矩阵的列=B矩阵的行   结果:A行B列
# 举例子              A矩阵: 2行3列   B:矩阵 3行2列   A3列=B3行           结果:2行2列 
```
